### Deliverable 1: Preprocessing the Data for a Neural Network

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
clean_df = application_df.drop(['EIN','NAME'], axis=1)
clean_df.head()
```

```python
# Determine the number of unique values in each column.
clean_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_counts = clean_df['APPLICATION_TYPE'].value_counts()
app_counts
```

```python
# Visualize the value counts of APPLICATION_TYPE
app_counts.plot.density()
```

```python
# Determine which values to replace if counts are less than 10
replace_application_indicator = clean_df['APPLICATION_TYPE'].value_counts() < 10
# switch the index with values 
replaced_application = pd.Series(replace_application_indicator.index.values, index=replace_application_indicator )
# select values to replace if counts are less than 10
replaced_application = replaced_application[replaced_application.index.values == True]

# Replace in dataframe
for app in replaced_application:
    clean_df.APPLICATION_TYPE = clean_df.APPLICATION_TYPE.replace(app,"Other")
    
# Check to make sure binning was successful
clean_df.APPLICATION_TYPE.value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
clean_df['CLASSIFICATION'].value_counts()
```

```python
# Visualize the value counts of CLASSIFICATION
clean_df['CLASSIFICATION'].value_counts().plot.density()
```

```python
# Determine which values to replace if counts are less than ..?
# less than 10
replace_class_indicator = clean_df['CLASSIFICATION'].value_counts() < 10
# switch the index with values 
replace_class = pd.Series(replace_class_indicator.index.values, index=replace_class_indicator )
# select values to replace if counts are less than 10
replace_class = replace_class[replace_class.index.values == True]

# Replace in dataframe
for cls in replace_class:
    clean_df.CLASSIFICATION = clean_df.CLASSIFICATION.replace(cls,"Other")
    
# Check to make sure binning was successful
clean_df.CLASSIFICATION.value_counts()
```

```python
# Generate our categorical variable lists
application_categories = clean_df.select_dtypes(include=['object']).columns.tolist()
application_categories
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(clean_df[application_categories]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(application_categories)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
application_df = clean_df.merge(encode_df, left_index=True, right_index=True)
application_df.drop(application_categories, 1, inplace=True)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df["IS_SUCCESSFUL"].values
X = application_df.drop(["IS_SUCCESSFUL"], axis=1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Deliverable 2: Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
no_of_features = len(X_train[0])
no_of_nodes_layer1 = int(no_of_features * 0.50)
no_of_nodes_layer2 = int(no_of_features * 0.25)
no_of_output_nodes = 1

### print all the layers input
print(no_of_features)
print(no_of_nodes_layer1)
print(no_of_nodes_layer2)

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=no_of_nodes_layer1, activation='relu', input_dim=no_of_features))


# Second hidden layer
nn.add(tf.keras.layers.Dense(units=no_of_nodes_layer2, activation='relu'))


# Output layer
nn.add(tf.keras.layers.Dense(units=no_of_output_nodes, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
from keras.callbacks import ModelCheckpoint
checkpoint = ModelCheckpoint("AlphabetSoupCharity.h5", verbose=1, save_weights_only=True, save_freq=5)
```

```python
# Compile the model
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model = nn.fit(X_train_scaled, y_train, epochs=50,verbose=1, validation_split=0.2, callbacks=[checkpoint])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
